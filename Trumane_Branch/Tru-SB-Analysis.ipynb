```python
import os
import csv
import pandas as pd

csvfile = pd.read_csv("Resources/directory.csv")

csvfile
```

```python
csvfile.columns
```

```python

cleaned_df = csvfile[{"Store Name","Ownership Type","Street Address","City","State/Province","Country","Postcode"}]
cleaned_df.head()
```

```python
new_df = cleaned_df.loc[cleaned_df["Country"] == "US", :]
new_df
```

```python
new_df['Zip_Code'] = new_df['Postcode'].where(new_df['Postcode'].str.len() == 5, 
                                               new_df['Postcode'].str[:5])
```

```python
new_df
```

```python
sb_df = new_df[{"Store Name","Ownership Type","Street Address","City","State/Province","Country","Zip_Code"}]
```

```python
sb_df
```

```python
incomefile = pd.read_csv("Resources/reduced_income_df.csv")
incomefile
```

```python
incomefile.dtypes
```

```python
incomefile['Zip_Code'] = incomefile['Zip_Code'].apply(str)
```

```python
merge_df = pd.merge(sb_df, incomefile, on="Zip_Code", how="inner")
merge_df
```

```python
result_df = merge_df.drop_duplicates(["Store Name"], keep='last')
```

```python
result_df
```
