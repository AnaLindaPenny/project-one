```python
import pandas as pd
```

```python
csvfile = "Resources/directory.csv"
starbucks_data = pd.read_csv(csvfile, encoding='utf-8')
starbucks_data.head()
```

```python
starbies_df = starbucks_data[["Store Number", "Store Name", "City", "State/Province", "Country", "Postcode", "Longitude", "Latitude"]]
starbies_df.head()
```

```python
starbucks_df = starbies_df.loc[starbies_df["Country"] == "US", :]
starbucks_df
```

```python
starbucks_df.to_csv("starbucks.csv", index=False, header=True)
```
